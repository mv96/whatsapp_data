```python
import numpy as np
import pandas as pd

file=input("enter the file that you want to open:")
fhand=open(file)

column_name=["date","time","sender","message"]
labels=[]
count=0

for rows in fhand:
    labels.append(count)
    count+=1
fhand.close()   

df=pd.DataFrame(columns=column_name,index=labels)
del(count,labels)
df.at[4, 'time']=4
#display(df)

fhand=open(file)

row=0
time_f=None #<<<< for some data sets turing this can save you
for line in fhand:
    line=line.strip()
    line=line.split("-")
    time_stamp=line[0]
    time=time_stamp.split(",")
    
    #print(time)
    type_1_length=(len(time[0])==len("24/04/17"))
    type_2_length=(len(time[0])==len("4/04/17"))
    type_3_length=(len(time[0])==len("4/4/17"))
    type_1_position=(time[0].find("/")==2)
    type_2_position=(time[0].find("/")==1)
    if((type_1_length or type_2_length or type_3_length) and (type_1_position or type_2_position)): #if in proper date format
        
        date=time[0]
        #print("in first if")
        time[1]=time[1].strip()#remove extra space
        if(len(time[1])==len("12:44 pm")):
            time_f=time[1]
            #print(time)
        elif(len(time[1])==len("1:54 pm")):
            time_f="0"+time[1]
            #print(time)
        elif(len(time[1])==len("1:54")):
            time_f="0"+time[1]
        elif(len(time[1])==len("12:54")):
            time_f=time[1]
   
        try:
            body=line[1]
            body=body.strip()
            mark=body.find(":")
            sender=body[:mark]
            if("secur"not in sender):
                sender=body[:mark]
                #print(sender)
            else:
                sender=np.nan
            message=body[mark+1:]
            message=message.strip()
            #print(sender)
        except:
            message=str(line[0])
            #>>>>>add message to data frame
    else:
        #print("in first else")
        #print("this is the message:",time)
        message=time
        
    df.at[row, 'date']=date
    df.at[row, 'time']=time_f
    df.at[row, 'sender']=sender
    df.at[row, 'message']=message
    row+=1

shape=df.shape

print("exchanges:{}, \nparameters:{}".format(shape[0],shape[1]))
inp=input("do you want to see the file (y/n) :")
if(inp=="yes"or inp=='y'):
    display(df)
inp=input("do you want to save the file in excel (y/n):")
if(inp=="yes"or inp=="y"):
    file_name=file[:file.find(".")]
    df.to_excel(file_name+".xlsx")
inp=input("do you also want to save the file in csv (y/n):")
if(inp=="yes"or inp=="y"):
    df.to_csv(file_name+".csv")
print("saved")

```
